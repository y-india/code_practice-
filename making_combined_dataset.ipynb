```python
import pandas as pd
import numpy as np


trades = pd.read_csv("historical_data.csv")
fg = pd.read_csv("fear_greed_index.csv")
```

```python
# Convert Fear Greed date to datetime and collapse extremes
fg['date'] = pd.to_datetime(fg['date']).dt.date

fg['sentiment'] = fg['classification'].replace({
    "Extreme Fear": "Fear",
    "Fear": "Fear",
    "Neutral": "Neutral",
    "Greed": "Greed",
    "Extreme Greed": "Greed"
})

# Keeping only relevant columns
fg = fg[['date', 'sentiment']]
```

```python
# Convert trade timestamp to date
trades['Timestamp'] = pd.to_datetime(trades['Timestamp IST'], dayfirst=True)
trades['date'] = trades['Timestamp'].dt.date

# Join sentiment
trades = trades.merge(fg, on='date', how='left')

# Drop rows with missing sentiment
trades = trades.dropna(subset=['sentiment'])

# Dataset 1: Clean trades with sentiment
clean_trades = trades.copy()
clean_trades.to_csv("clean_trades.csv", index=False)
```

#### Dataset 1: Clean Trades with Sentiment

- Converted trade timestamps to date for daily aggregation.

- Collapsed sentiment categories: Extreme Fear → Fear and Extreme Greed → Greed.

- Merged trades with Fear Greed index on date.

- Dropped trades without a matching sentiment.

- Result: Every trade now has a valid sentiment label, giving context for behavior analysis.

making clean dataset of dataset 2

```python
# Define order-level grouping
order_cols = ['Account', 'Order ID', 'Coin', 'Side', 'date', 'sentiment']

orders = clean_trades.groupby(order_cols).agg(
    order_notional_usd=('Size USD', 'sum'),
    order_fee=('Fee', 'sum'),
    fills=('Trade ID', 'count'),
    crossed_pct=('Crossed', lambda x: np.mean(x == True)),
    net_pnl=('Closed PnL', 'sum')
).reset_index()
```

```python
# Group by trader, date, sentiment
trader_day = orders.groupby(['Account', 'date', 'sentiment']).agg(
    daily_net_pnl=('net_pnl', 'sum'),
    total_fees=('order_fee', 'sum'),
    num_orders=('Order ID', 'count'),
    total_usd_traded=('order_notional_usd', 'sum'),
    avg_crossed_pct=('crossed_pct', 'mean')
).reset_index()

# Dataset 2: Trader day table

trader_day.to_csv("clean_trader_day.csv", index=False)
```

#### Dataset 2: Trader Day Dataset

- Aggregated trades into order level:

     - Summed USD size, fees, Net PnL

     - Counted fills

    - Calculated percentage of crossed orders

- Aggregated orders into trader day level:

     - Daily Net PnL

     - Total fees

    - Total USD traded

     - Number of orders

    - Average crossed percent

- Result: Each row represents one trader on one day under one sentiment, ready for analysis of behavior and performance.
